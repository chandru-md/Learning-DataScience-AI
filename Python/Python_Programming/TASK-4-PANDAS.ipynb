### REMOVING LEADING AND TRAILING SPACES

```python
def trim_whitespaces(sales_data):

    for x in sales_data.columns:
        if sales_data[x].dtype == 'Object':
            sales_data = sales_data[x].str.strip()
    return sales_data
```

```python
sales_data = trim_whitespaces(sales_data)
```
