### ROCK PAPER SCISSORS GAME

```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

game_images = [rock, paper, scissors]

try:
    user_choice = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.\n"))
    if user_choice < 0 or user_choice > 2:
        print("You typed an invalid number. pLease enter a number between 0 to 2")
    else:
        print(game_images[user_choice])

        computer_choice = random.randint(0, 2)
        print("Computer chose:")
        print(game_images[computer_choice])
        
        
        if user_choice == 0 and computer_choice == 2:
            print("You win!")
        elif computer_choice == 0 and user_choice == 2:
            print("You lose!")
        elif computer_choice > user_choice:
            print("You lose!")
        elif user_choice > computer_choice:
            print("You win!")
        elif computer_choice == user_choice:
            print("It's a draw!")

except ValueError:
    print("You typed an invalid number. Please enter a number between 0 to 2")
```

### FOR LOOP

```python
# FOR LOOP WITH LIST

dogs = ['sam','jimmy','tom']

for dog in dogs:
    print(dog)
    print(dogs)
```

```python
# FOR LOOP WITH RANGE()

for i in range(1,11):
    print(i)
```

```python
# for loop using a string
for x in "CHANDRU":
    print(x)
```

```python
# FOR LOOP USING ENUMERATE

names = ["Chandru","Kathiravan","Deivanayagan"]

for index,name in enumerate(names):
    print(index,name)
```

```python
# FOR LOOP USING DICTIONARY

names = {"name1":"Chandru","name2":"Kathiravan","name3":"Deivanayagan",
        "age1":23,'age2':31,'age':56,
         "occupation1":"unemployed",'occupation2':"Manager","occupation3":"Public sector"
        }

for key,value in names.items():
    print(key,'----->',value)

```
