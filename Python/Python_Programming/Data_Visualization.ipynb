##### Bar plot

```python
plt.bar(data=mt_cars_data,x="am",height="mpg")
plt.show()
```

```python
mt_cars_data['cyl'].value_counts()
```

```python
plt.pie(x=mt_cars_data["cyl"].value_counts(),labels=["8 Cyl","4 Cyl","6 Cyl"],explode=[0.01,0.01,0.01],autopct='%1.0f%%')
plt.figure(figsize=(10,15))
plt.show()
```

```python
mt_cars_data.corr()
```

```python
corr_matrix = mt_cars_data.corr()
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.figure(figsize=(10,15))
```

#### notes

- 0 to 0.20 ---- very week correlation
- 0.21 to 0.40 ---- week correlation
- 0.41 to 0.60 ---- standard / normal correlation
- 0.61 to 0.80 ---- strong correlation
- 0.81 to 1.0 ----- very strong correlation

## Advanced Data Visualization

#### STATIC CHARTS - MUTABLE, DYNAMIC CHARTS - IMMUTABLE

```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```
