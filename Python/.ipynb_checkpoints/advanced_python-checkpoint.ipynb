#### parameters and arguments

```python
# UDF ---- USER DEFINED FUNCTIONS
```

```python
def get_details(name,place,excepted_price):
    print("Client Name:",name)
    print("Place:",place)
    print("Excepted Price:",excepted_price)
```

#### positional arguments

```python
get_details("Sweety Omana","Kanyakumari",'100M')
```

```python
get_details(excepted_price=99.45,name='Chandru',place='Mumbai')
```

#### default arguments

```python
def get_details(name,excepted_price,place = 'Chennai'):
    print("Client Name:",name)
    print("Place:",place)
    print("Excepted Price:",excepted_price)
```

```python
get_details('Kathiravan',105.67)
```

#### VARIABLE LENGTH ARGUMENTS

```python
def get_details(name,place,*excepted_price):   ## get Multiples input**
    print("Client Name:",name)
    print("Place:",place)
    print("Excepted Price:",excepted_price)
```

```python
get_details("Deivanayagan","Neyveli",95.9,1005,105.26)
```

#### keyword arguments

```python
def get_details(name,place,**excepted_price):   ## get Multiples input with keys**
    print("Client Name:",name)
    print("Place:",place)
    print("Excepted Price:",excepted_price)
```

```python
get_details("Mayavan","India",ask1=90,ask2=89,ask5=900,ask3=100)
```

```python
def get_details(name,*place,**excepted_price):   ## get Multiples input with keys**
    print("Client Name:",name)  ## both variable length and keyword arguments
    print("Place:",place)
    print("Excepted Price:",excepted_price)
```

```python
get_details("Chandru","Russia","Switzerland",ask1=1000,ask2=2000)
```

#### control the flow of the program

```python
user_passcode = input("New passcode: ")
if len(user_passcode)<10:
    print('week passcode')
else:
    print('Strong passcode')
```

```python
user_passcode = input("New passcode: ")
if len(user_passcode)<10:
    print('week passcode')
elif 15<=len(user_passcode)<=20:
    print("Moderate Passcode")
else:
    print('Strong passcode')
```

#### find the factors of a number

```python
user_input = int(input("Enter your Number: "))
for i in range(1,user_input+1):
    if user_input%i==0:
        print(i)
```

```python
user_input= int(input("Enter your number: "))

result = []

for i in range(1,user_input+1):
    if user_input%i==0:
        result.append(i)
print("The factors of {} is {}".format(user_input,result))
    
```
