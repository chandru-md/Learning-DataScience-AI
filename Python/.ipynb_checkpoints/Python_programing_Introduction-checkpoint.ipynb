```python
print("Welcome to number guessing name!!!!")
print("Guess the Number between 1 to 100")
secret_number = 56
for i in range(1,8):
    user_guess = int(input("Enter the number: "))
    if user_guess<1 or user_guess>100:
        print("Invalid number! please enter the number between 1 to 100.")
    elif user_guess > secret_number:
        print("your number is higher")
    elif user_guess < secret_number:
        print("your number is lesser")
    else:
        print("Congratulations, you guessed the correct number!")
        print(f"You have guessed the number {secret_number} in {i} attempts")
        break
```

```python
import random

print("Welcome to number guessing name!!!!")
print("Guess the Number between 1 to 100")
secret_number = random.randint(1,100)
for i in range(1,8):
    try:
        user_guess = int(input("Enter the number: "))
        if user_guess<1 or user_guess>100:
            print("Invalid number! please enter the number between 1 to 100.")
        elif user_guess > secret_number:
            print("your number is higher")
        elif user_guess < secret_number:
            print("your number is lesser")
        else:
            print("Congratulations, you guessed the correct number!")
            break
    except ValueError:
        print("Invalid characters are not recognized. please enter the numbers between 1 to 100.")
if user_guess == secret_number:
    print(f"You have guessed the number {secret_number} in {i} attempts")
else:
    print("Sorry! All trials is exceeded. Try again!!")
```

```python
def number_guessing_game():
    import random

    print("Welcome to number guessing name!!!!")
    print("Guess the Number between 1 to 100")
    secret_number = random.randint(1,100)
    for i in range(1,8):
        try:
            user_guess = int(input("Enter the number: "))
            if user_guess<1 or user_guess>100:
                print("Invalid number! please enter the number between 1 to 100.")
            elif user_guess > secret_number:
                print("your number is higher")
            elif user_guess < secret_number:
                print("your number is lesser")
            else:
                print("Congratulations, you guessed the correct number!")
                break
        except ValueError:
            print("Invalid characters are not recognized. please enter the numbers between 1 to 100.")
    if user_guess == secret_number:
        print(f"You have guessed the number {secret_number} in {i} attempts")
    else:
        print("Sorry! All trials is exceeded. Try again!!")
```

```python
number_guessing_game()
```
